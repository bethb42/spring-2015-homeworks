## Step 3. Advanced filtering & joins

By now, you should have noticed that, for each day, there are multiple records in our dataframe. Use the `groupby` function to get, for each day, the counts for all the case categories. Plot the result for the top 5 categories as you see fit. **(20 pts)**

````python
# Write your code here
five_common = ['Request for Snow Plowing', 'Snow Removal', 'Requests for Street Cleaning','Request for Pothole Repair',\
'Misc. Snow Complaint' ]
date_list = []

#makes start and end dates so I can premake the entire dictionary
start = "01-01-2015"
start_day = datetime.strptime(start, "%m-%d-%Y")
end = "02-15-2015"
end_day = datetime.strptime(end, "%m-%d-%Y")
# one_day = datetime.timedelta(days=1)

#making a dictionary of dictionary
#this will hold the results of the days and the requests
common_dict = {}
for case in five_common:
    date_dict = {}
    temp_day =  start_day

    while(temp_day <= end_day):
        date_dict[temp_day.date()] = 0
        temp_day = temp_day + timedelta(days=1)
        
    common_dict[case] = date_dict

    

boston.index = boston.OPEN_DT

#goes through and counts the number of requests per day for each of the most common complaints. 
for DAY, dates_dt in boston.groupby(pd.TimeGrouper(freq = "1D")):
    
    for CASE, case_dt in dates_dt.groupby("CASE_TITLE"):
        
        if CASE in five_common:
            common_dict[CASE][DAY.date()] =  len(case_dt.CASE_TITLE)

dd = pd.DataFrame(common_dict)
dd.plot()
````

Let's see how this trend compares to the snowfall. First, we need to read the data from `snow.csv`

````python
# Read the dataframe
snow_data = pd.read_csv("snow.csv",parse_dates=[1])
````

According to the documentation, the column `SNOW` has the values for the snowfall, in mm. However, the dataset we have has information from many different stations around Boston. Aggregate all their measurements and transform the result to centimeters. Finally, plot the result **(10 pts)**

````python

#removing the invalid values
for snow_value in snow_data.DATE:
    if snow_value == -9999:
        snow_value = 0
        


snow_dict = {}

for day, snow_dates_dt in snow_data.groupby(snow_data.DATE):
    #find the max for a particular day and dividing it by 10 to get mm
    snow_dict[day.date()] = snow_dates_dt.SNOW.max() /10

dd_snow = pd.DataFrame(snow_dict.items(), columns=['Date', 'SNOW'])
dd_snow = pd.Series(snow_dict, name='DateValue')
#plotting
dd_snow.plot()
````

## Step 4. Matching trends

The last thing we will do is to join the snowfall dataframe with that of the top 5 categories of calls, per day. Don't forget to fill the NaN values with 0.
````python
calls_on_snow = dd
calls_on_snow['Snow'] = dd_snow  #snow was a series, so I can just add another column
# calls_on_snow
````

Now, you can show the connection of the two trends (the calls and the snow) in a plot. I provide some sample code, but you should also try your own versions.

````python
fig, axes = plt.subplots(nrows=2,figsize=(12,5), sharex=True)

calls_on_snow.ix[:,0:5].plot(ax=axes[0])
calls_on_snow.ix[:,5].plot(ax=axes[1],label="snowfall (in cm)", color='gray')

axes[0].set_title("The effect of the winter snowstorms on Boston's 311 Hotline for 2015")
axes[0].set_ylabel("# Calls")
axes[1].set_ylabel("Snowfall in cm")
axes[1].set_xlabel("Date")
axes[0].xaxis.grid(False)
axes[1].xaxis.grid(False)
````
