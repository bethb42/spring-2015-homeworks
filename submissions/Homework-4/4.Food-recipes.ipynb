```python
import time

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.datasets as datasets
import sklearn.metrics as metrics
import sklearn.utils as utils


import scipy.sparse.linalg as linalg
import scipy.cluster.hierarchy as hr

import sklearn.cluster as cluster

#classification
from sklearn.neighbors import KNeighborsClassifier


from sklearn.decomposition import TruncatedSVD
from scipy.spatial.distance import pdist, squareform
from sklearn.preprocessing import StandardScaler

import seaborn as sns
%matplotlib inline
print "done"
```

```python
#importing the file as a text file
file_obj =  open('allr_recipes.txt', 'r')
#dict food will hold the country and the number of times each word in it occurs. 
dict_food = {} 
#count holds the country and how many recipies there are from that country
count = {}

#go through the file object to make the above two 
for line in file_obj:
    lst = line.split()
    ingredients = lst[1:]
    country = lst[0]
    #combines and counts each countries ingredients. 
    if country not in dict_food:
        dict_food[country] = {}
    for word in ingredients:
        if word in dict_food[country]:
            dict_food[country][word] +=1
        else:
            dict_food[country][word] = 1
            
    #counts how many recipies there for each country
    if country not in count:
        count[country] = 0
    count[country] += 1

#makes dict_food into a dataframe and makes all Nan equal to 0. 
db_food =pd.DataFrame(dict_food)
db_food =db_food.fillna(value = 0)
countries = db_food.columns
countries = list(countries)

#transverses the table to that each country is a row; will make clustering easier. 
db_food = db_food.T
db_food

#the dataframe now holds the count of how many times each word appeared in recipies labeled
#with their country. 
```

```python
#Go through the dataframe and divide the number of times a word appeared by the 
#number of recipies in order to get what percent of the time each word appears. 
#This will help offsest that some countries have a lot more recipies than others. 
i =0
for idx, row in db_food.iterrows():
    for j in range(len(row)):
        row[j] = row[j]/count[countries[i]]
    i += 1
#     print row[0]
db_food
count
```

```python
#This clusters the different countries. 
distanceMatrix = pdist(db_food,'euclidean')
Z = hr.linkage(distanceMatrix, method='complete', metric='euclidean')
fig = plt.figure(figsize=(15,20))
T = hr.dendrogram(Z,color_threshold=1, leaf_font_size=12, labels = countries) 
```

My hypothesis when I first saw this data was that countries that are geographically near each other or are culturally related would have more similar foods. In counted how many times each recipie item appeared in the list a country's recipe list.

I then divided this number to  get what percent each particular food is of the overall foods for that country. 

I then did hierarchical clustering to see how close various countries are to each other. 

Some of the results made a lot of sense. America and Canada have very similar food. Spain, Portugal, the Caribbean, and South America were clustered closely. However, some of the results were a little unclear. Mexico was shown to be most similar to South-Africa instead of South America, which is unexpected because they are neither geographically closer nor culturally related. 

Bangladesh and Pakistan were clustered closely together, which also makes a fair amount of sense. 

Israel was combined with a large group of mostly European countries, likely reflecting the fact that many Jews from Europe moved to Israel after World War II. 

These results show that for the most part, my hypothesis was correct. The countries that don't fit into my hypothesis may due to the sample size. While America had thirty-five thousand recipes, some countries had as few as four recipes. With many of these countries, it may not have been possible to get a good representative of the foods that typify that place.
